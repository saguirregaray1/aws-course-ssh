```python
#Convert columns to snake_case

apps.rename(lambda x: x.lower().strip().replace(' ', '_'), 
            axis='columns', inplace=True)
```

```python
apps.columns
```

```python
# Specify the cols to drop
to_drop = ['installs','minimum_installs','minimum_android', 'free', 'price', 'currency', 'size', 'released',
    'developer_id', 'developer_website', 'developer_email', 'privacy_policy', 
     'in_app_purchases', 'editors_choice', 'scraped_time'
]

# Drop them
apps.drop(to_drop, axis='columns', inplace=True)
```

```python
apps.columns
```

```python
# Collapse 'Music' and 'Music & Audio' into 'Music'
apps['category'] = apps['category'].str.replace('Music & Audio', 'Music')
```

```python
# Collapse 'Educational' and 'Education' into 'Education'
apps['category'] = apps['category'].str.replace('Educational', 'Education')
```

```python
#Buscamos cuales son las 5 categorías con más aplicaciones

sqldf("select COUNT(*),category from apps group by category order by COUNT(*) DESC LIMIT 5",globals())
```

```python
top_5_list = [
    'Education', 'Music', 'Business', 'Tools', 
    'Entertainment'
]

top = apps[apps['category'].isin(top_5_list)].reset_index(drop=True)
```

```python
top['last_updated'] = pd.to_datetime(top['last_updated'], format='%b %d, %Y',
                                 infer_datetime_format=True)
```

```python
top['content_rating'].value_counts()
```

```python
#Colapsando los content_ratings para tener solamente Everyone, Teen, Adults

top['content_rating'] = top['content_rating'].replace('Unrated',"Everyone")
top['content_rating'] = top['content_rating'].replace('Mature 17+',"Adults")
top['content_rating'] = top['content_rating'].replace('Adults only 18+',"Adults")
top['content_rating'] = top['content_rating'].replace('Everyone 10+',"Everyone")
```

```python
top['content_rating'].value_counts()
```

```python
#Agregamos el feature market_segment concatenando category y content rating

top['market_segment'] = top['category'] + ' | ' + top['content_rating']
```

```python
top['market_segment'].value_counts()
```

```python
top.head()
```

```python
#Filtramos todas las apps que no sean ad_supported
top = top[top['ad_supported'] == True]
```

```python
 #Filtramos todas las apps que no hayan sido actualizados desde 2020
top = top[top['last_updated'] >= '2020-01-01']
```
